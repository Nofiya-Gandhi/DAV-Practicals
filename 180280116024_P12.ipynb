## Dataset



```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data

Data = pd.read_csv('StoreData.csv')
```

```python
Data.head()
```

## Visualize data

```python
#plot data
# create line plot
sns.lineplot(x='Month', y='S1', data=Data, color='blue')
sns.lineplot(x='Month', y='S2', data=Data, color='red')
sns.lineplot(x='Month', y='S3', data=Data, color='green')
sns.lineplot(x='Month', y='S4', data=Data, color='yellow')
sns.lineplot(x='Month', y='S5', data=Data, color='black')
plt.title('Monthly Profit of different Stores')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.xticks(range(1,13))
plt.show()
```
